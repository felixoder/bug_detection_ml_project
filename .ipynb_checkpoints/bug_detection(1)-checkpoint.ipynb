```python
csv_filename = "github_buggy_code.csv"
```

```python
# Load the CSV file into a pandas DataFrame
try:
    df = pd.read_csv(csv_filename)
except FileNotFoundError:
    df = pd.DataFrame({"code": [], "label": []})
```

```python
# Split the dataset into training and validation sets
train_df, val_df = train_test_split(df, test_size=0.1)  # 10% for validation
```

```python

# Convert the train and validation DataFrames into Hugging Face datasets
train_dataset = Dataset.from_pandas(train_df)
val_dataset = Dataset.from_pandas(val_df)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from datasets import DatasetDict

# Load pre-trained model and tokenizer
MODEL_NAME = "microsoft/codebert-base"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=2)

# Inspect the dataset columns
print(train_dataset.column_names)
print(val_dataset.column_names)

# Check unique labels in the dataset to update label mapping
print(set(train_dataset['label']))

# Function to tokenize the code (adjust the column name as necessary)
def tokenize_function(examples):
    return tokenizer(examples['code'], padding=True, truncation=True, max_length=512)  # Replace 'code' with correct column name

# Apply tokenization to both train and validation datasets
train_dataset = train_dataset.map(tokenize_function, batched=True)
val_dataset = val_dataset.map(tokenize_function, batched=True)

# Update the label map according to your actual dataset labels
label_map = {'bug-free': 0, 'buggy': 1}  # Update this based on actual labels in your dataset

# Apply the label mapping
train_dataset = train_dataset.map(lambda x: {'label': label_map[x['label']]})
val_dataset = val_dataset.map(lambda x: {'label': label_map[x['label']]})

# Now create a DatasetDict with both 'train' and 'validation' splits
dataset = DatasetDict({
    "train": train_dataset,
    "validation": val_dataset
})
```

```python
from sklearn.metrics import accuracy_score
from transformers import Trainer, TrainingArguments

# Define compute_metrics function to calculate accuracy
def compute_metrics(p):
    predictions, labels = p
    preds = predictions.argmax(axis=1)  # Convert logits to predicted class labels
    return {"eval_accuracy": accuracy_score(labels, preds)}

def train_model():
    # Adjusted training arguments
    training_args = TrainingArguments(
        output_dir="./results",
        eval_strategy="epoch",  # Use 'eval_strategy' instead of deprecated 'evaluation_strategy'
        save_strategy="epoch",
        logging_strategy="epoch",
        logging_steps=10,
        learning_rate=1e-6,  # Further reduce learning rate
        per_device_train_batch_size=8,
        num_train_epochs=5,  # Continue with the number of epochs
        weight_decay=0.01,
        warmup_steps=500,  # Adding a warm-up period to stabilize training
        load_best_model_at_end=True,  # Optionally load the best model at the end of training
        metric_for_best_model="eval_accuracy",  # Track accuracy to select the best model
        greater_is_better=True  # Ensure better performance is selected as the best model
    )

    # Trainer with compute_metrics function
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
        tokenizer=tokenizer,  # Tokenizer is used here, replace it when upgrading to the new version
        compute_metrics=compute_metrics  # Add this line to calculate accuracy during evaluation
    )

    trainer.train()
```

```python
# Train the model
train_model()
```
